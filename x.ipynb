```python
from qiskit import *
from numpy import *
from qiskit.visualization import plot_histogram


```

```python
#eintein's measurements
#x
n=4
circuit1=QuantumCircuit(n,n)
circuit1.h(0)
circuit1.cx(0,1)
circuit1.cx(0,2)
circuit1.x(0)
circuit1.cx(1,3)
circuit1.measure(0,0)
%matplotlib inline
circuit1.draw(output='mpl')
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(circuit1,backend=simulator,shots=4096).result()
plot_histogram(result.get_counts(circuit1))
```
